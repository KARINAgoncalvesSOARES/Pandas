## Como filtara strings no pandas:

[Python PANDAS - A melhor forma de filtrar dados string](https://www.youtube.com/watch?v=-1EM45iFe_8&list=PLEdfCNsWt0gxURVswsV_STMrCo_n7rD5M&index=4&ab_channel=ProgramePython)

```python
import pandas as pd
```

```python
dados = [['ola', 'bem', 'vindos'], ['ao','programe', 'python+'],['aula', 'de', 'hoje'], ['filtros', 'pandas', 'python']]
```

```python
df = pd.DataFrame(dados, columns=['col1','col2','col3'])
df
```

```python
#Primeira condição

df[df['col1'] == 'ola']
```

```python
#.isin(), filtra apenas a linha em que está a palavra que for indicada.

df[df['col3'].isin(['python', 'python+'])]
```

```python
#str, há uma infinidade  de funções onde podemos a usar(consultar documentação)
#mas aqui nós vamos fazer contagem da quantidade de caracteres que uma string possui e passa uma condição para a impressão:

df[df['col3'].str.len() > 5]
```

```python
#.str.contains(), proucura strings no meu data frame que contenham por ex a a sílaba indicada

df[df['col3'].str.contains('os|e|\+')]
```
