```python
import pandas as pd 
import numpy as np
```

```python
dados = [[111.23548, 478.245784],
         [222.58485, 444.888978],
         [123.456578, 555.585478]]
        
dados1 = [[111.23548e+05, 478.245784e+00],
         [222.58485e+06, 444.888978e+06],
         [123.456578e+06, 555.585478e+05]]      
```

```python
df = pd.DataFrame(dados,columns=('col1','col2'))
df1 = pd.DataFrame(dados1, columns=('col1','col2'))
```
