## São exibidas 5 linhas por default, mas é possivel mostrar mais linhas passando a quantidade desejada como padrão.

```python
arquivo.head(3)
```

```python
arquivo.tail()
```

##2- dataframe.info(), nos traz um resumo preciso do dataframe

```python
arquivo.info()
```

##3- Shape nos mostra a quantidade de linhas e a quantidade de colunas que o DataFrame possui.

```python
arquivo.shape
```

##4- Visualizando os tipos dos dados de cada coluna.

```python
arquivo.dtypes
```

##5- Retorna o tamanho do DataFrame que é a quantidade de linhas x quantidade de colunas.

```python
arquivo.size
```

##6- O método describe(), mostra um resumo estatístico de cada coluna númerica do DataFrame.

```python
arquivo.describe()
```

## Agora eu estou usando o operador de transposição para mudar a posição das colunas para linhas.

```python
arquivo.describe(include='all').T
```

##7- O método sample() nos permite visualizar valores aleatórios do DataFrame ou de Series.

```python
arquivo.sample(n=3)
```

##8- Identificador de valores missing(NaN).
## Informa se temos algum valor missing o identificando como True, caso o DataFrame não possua valor NaN ele retorna como False. Os valores missing tem efeito significativo na análise de dados.

```python
arquivo.isnull()
```

##9- Isna possui a mesma função do isnull, retornar se o DataFrame possui valores nulos, retornando valores boleanos com True indicando valores missing.

```python
arquivo.isna().any()
```

##10- Identificando a soma de valores nulos(missing).Você pode calcula o número de valores ausentes em cada coluna.

```python
arquivo.isnull().sum()
```
