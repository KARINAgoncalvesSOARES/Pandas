#Função Groupby

```python
# É usada para dividir os dados em grupos com base em alguns critérios.

arquivo[['housing_median_age','median_house_value','total_rooms']].groupby(['housing_median_age','median_house_value']) .mean()
```

# Sorting

##classificando o índice do DataFrame

```python
# Por padrão, ele classifica em ordem crescente, para classificar em ordem decrescente, use ascending=False

arquivo.sort_index(axis=1, ascending= True)
```

# Sort by values

```python
#A função sort_values() serve para ordenar o seu dataframe a partir de uma coluna, 
#seja em ordem crescente de maneira padrão ou decrescente, passando o argumento ascending = False .

arquivo.sort_values(by='median_house_value')
```

# nsmallest() e nlargest()

Suponho que não haja dúvidas sobre os propósitos desses dois métodos depois de apenas ler seus nomes, mas, no entanto, eles podem ser valiosos no processo de análise exploratória de dados.

Vamos ver como faríamos para encontrar as 5 observações com o menor valor

```python
arquivo.nsmallest(5,'median_house_value')
```

Vamos ver como faríamos para encontrar as 5 observações com o maior valor.

```python
arquivo.nsmallest(5,'median_income')
```

# Drop

A função dropna() é usada para remover uma linha ou coluna de um dataframe que tenha NaN ou nenhum valor nele .

```python
drop_cols = ['ocean_proximity']
arquivo.drop(drop_cols,axis=1,inplace=True)
arquivo
```

# Query

Às vezes, precisamos filtrar um dataframe com base em uma condição ou aplicar uma máscara para obter determinados valores. Uma maneira fácil de filtrar um dataframe é a função de consulta . Vamos primeiro criar um dataframe de amostra.



```python
arquivo.query('5000 < total_rooms < 5500')[:5]
```

```python
arquivo.query('40000 < median_house_value < 100000')[:8]
```

# Insert column

Quando queremos adicionar uma nova coluna a um dataframe, ela é adicionada no final por padrão. No entanto, os pandas oferecem a opção de adicionar uma nova coluna em qualquer posição usando a função de inserção .

```python
# nova coluna 
new_col=np.random.randn(20640)
#insira a nova coluna na posição 8
arquivo.insert (8,'new_col',new_col)
```
