##6- O método describe(), mostra um resumo estatístico de cada coluna númerica do DataFrame.

##8- Identificador de valores missing(NaN).
## Informa se temos algum valor missing o identificando como True, caso o DataFrame não possua valor NaN ele retorna como False. Os valores missing tem efeito significativo na análise de dados.

```python
arquivo.isnull()
```

##9- Isna possui a mesma função do isnull, retornar se o DataFrame possui valores nulos, retornando valores boleanos com True indicando valores missing.

```python
arquivo.isna().any()
```

##10- Identificando a soma de valores nulos(missing).Você pode calcula o número de valores ausentes em cada coluna.

```python
arquivo.isnull().sum()
```

#Loc e iloc

```python
# Loc é usado para selecionar por rótulos
# O rótulo da coluna ou nomes de colunas

arquivo.loc[:5,['median_income','median_house_value']]
```

```python
# iloc é usado para selecionar por posição
# usado para selecionar linhas

arquivo.iloc[:4, :6]
```

# Slicing/fatiamento

```python
arquivo[0:4]
```

#Nunique

```python
# Nunique conta o número de entradas únicas nas colunas (ou) linhas
# é muito útil em recursos categóricos.

arquivo.nunique()
```

#Memory usage

```python
# Retorna o uso de memória de cada coluna em bytes.

arquivo.memory_usage()
```

#Sample

```python
# Gera amostras aleatórias de cada grupo de um objeto DataFrame

arquivo.sample(n=8)
```

#Função Groupby

```python
# É usada para dividir os dados em grupos com base em alguns critérios.

arquivo[['housing_median_age','median_house_value','total_rooms']].groupby(['housing_median_age','median_house_value']) .mean()
```

```python
arquivo[['median_house_value','housing_median_age']].groupby('median_house_value').agg(['mean','count'])
```
