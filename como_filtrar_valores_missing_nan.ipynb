```python
import pandas as pd
import numpy as np
```

```python
dados = [[6.0, np.nan, 15.0 ], [np.nan, 15.0, np.nan], [16.0, 1.0, np.nan], [4.0 ,19.0, 16.0], [16, 8, 10], [12, 9, 11]]
```

```python
df = pd.DataFrame(dados, columns=['col1','col2','col3'])
df
```

```python
#.isnull(), etorna apenas as linhas que tem o valor missing(nan)
df[df['col3'].isnull()]
```

```python
#.notnull(), retorna apenas as linhas que N√ÉO possuem valor missing(nan)

df[df['col3'].notnull()]
```

```python
# o "~" nega a fun√ß√£o,no caso dando o mesmo objetivo do notnull.
df[~df['col3'].isnull()]
```

```python
#.isna() tem a mesma fun√ß√£o e a mesma documenta√ß√£o do .isnull()(retorna as linhas que possuem valor missing)

df[df['col3'].isna()]
```

```python
#.any() analisa tanto linha quanto coluna  conforme a instru√ß√£o anterior
#No nosso exemplo a intru√ß√£o anterior √© .isnull()

df[df.isnull().any(axis=1)]
```

```python
# Nessa pr√≥xima fun√ß√£o eu estou pedindo para o colab imprimir apenas as linhas que possuem mais de um NaN.
# Para que isso aconte√ßa √© necess√°rio somar a quantidade de missing que aparecem.

df[df.isnull().sum(axis=1) == 2]
```
