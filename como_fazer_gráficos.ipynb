```python
import pandas as pd
import numpy as np
```

```python
dados = [['karina', 63.0, 1.64],
         ['Eddy', 63.0, 1.51],
         ['Elza', 75.0, 1.51],
         ['Washington',66.0, 1.70]]
```

```python
arquivo= pd.DataFrame(dados, columns=('nomes', 'peso','idade'))
```

```python
arquivo.head()
```

```python
arquivo.plot()
```

```python
#kind --> tipo de gráfico que eu quero usar:
arquivo.plot(kind='bar')
```

image.png

```python
# x é útil quando eu quero um valor para o meu x  diferentes do me índice.
# y é quando eu quero os valores apenas de uma colula

arquivo.plot(y='idade', kind='bar')
```

```python
#subplots=True é uma forma diferente de imprimir os gráficos, o layout é uma tupla com o número de colunas e o número de linhas.
#figsize=() é a largura e a altura da impressão dos greficos
#sharey e sharex significa compartilhar o x e compartilahar o y
#sharey como resultado muda as plotagens dos resultador, permitindo ver com mais clareza quais são os gráficos que tem mais volume
arquivo.plot(kind="bar", subplots=True, layout=(2,2), figsize=(10,10),sharey=True)
```

```python
#Parametro stacked permite criar o grafico de barras de forma empilhada
arquivo.plot(kind='bar', stacked=True)
```
