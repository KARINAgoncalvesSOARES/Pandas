# Pacote Pandas

Outra biblioteca ou pacote que compõe o PyData Stack é o Python Data Analysis
Library (Pandas). O seu foco está em análise de dados, mas dependendo do caso pode
servir para a visualização de dados em formato de tabelas, de modo que podemos
preparar os dados que serão apresentados por meio de gráficos.

```python
import pandas as pd
import numpy as np
```

###Criando Series

```python
s1 = pd.Series([23, 77, 0, 5, 12])
print(s1)
print(' ')

print(s1.values) # Para verificar os valores serie
print(' ')

print(s1.index) # Para verificar os índices da serie
print(' ')
```

```python
# Criando serie e específicando os índices
s2 = pd.Series([23, 77, 0, 5, 12], index = ['A', 'B', 'C', 'D', 'E'])
print(s2)
print('\n')

print(s2[:2]) # Fatiamento de series
print('\n')

print(s2[s2>20]) # Retornando os elementos maiores que 20
print('\n')

print(s2['A']) #Trabalhando com posição específica da serie
# ou           #devemos específicar seu index label ou numérico    
print(s2[0])
print('\n')

s2['A'] = 100
print(s2)
```

```python
# Criando uma serie com valor NaN
s3 = pd.Series([23, 77, 0, 5, 12,np.NaN], index = ['A', 'B', 'C', 'D', 'E','F'])
print(s2)
print('\n')

print(pd.isnull(s3)) # Verificando se temos dados  nulos
print('\n')

print(pd.notnull(s3)) # Verificando se temos dados não nulos
```

#### Juntando serie
Perceba que quando concatenamos, os elementos com o mesmo índice são somados em valores e os elementos que são diferentes – neste caso, o índice F – são
adicionados como novos na junção.

```python
# Juntando series
s4 = s2 + s3
print(s4)
```

#Criando e manipulando Data Frames

Agora que já vimos como criar séries com o Pandas, conferiremos exemplos de
criação e manipulação de dataframes:

 Dataframes possuem index e linhas semelhantes a uma planilha.

```python
# Criando um dicionário
dados= {'Ano':[2014, 2015, 2016, 2017, 2018, 2019], 
        'Pais':['EUA', 'China', 'Brasil', 'Canada', 'Portugal', 'Espanha'],
        'Vendas':[50, 60, 70, 40, 10, 88]}
```

```python
# Criando DataFrame a partir do dicionário
df = pd.DataFrame(dados)
df
```

```python
# Adicionando colunas ao dicionário
df2 = pd.DataFrame(dados, columns = ['Ano', 'Pais', 'Vendas', 'Valores'],
                   index = ['A', 'B', 'C', 'D', 'E', 'F'])

df2
```

```python
# Manipulando apenas uma coluna do DataFrame
df2['Pais']
```

```python
# Também podemos atribuir uma coluna para uma variável
s = df2['Ano']

print(s)
print('\n')
print(type(s)) # Verificando o type de uma serie do dataframe
```

```python
print(df2.index) # Visualizando index 
print('\n')

print(df2.columns) # Visualizando Colunas
print('\n')

print(df2.values) # Visualizando os dados/values
print('\n')
```

```python
# Realizando fatiamento/slicing do data frame
df2[:2]
```

```python
# Visualizando o início dos dados
df2.head()
```

```python
# Visualizando o final dos dados
df2.tail()
```

#### Usando Pandas e Numpy

```python
# Usando Numpy para preencher valores vazios do nosso dataframe com valores randômicos

df2['Valores'] = np.random.rand(6)*100

df2
```

#### Um método bem interessante é o describe(), que descreve os dados do dataframe; este método retorna:
* **count** = número de registros (linhas)
* **mean** = média dos valores 
por coluna
* **std** = desvio padrão
* **min** = valor mínimo da amostra
* **quartis** = 25%, 50% e 75%
* **max** = valor máximo da amostra

```python
# describe()
df2.describe()
```

```python
# Fatiamento com índece
print(df2['A':'C'])
print('\n')

print(df2[0:3])
```
